```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from collections import Counter
nltk.download('punkt')
def getText():
    txt = open("hamlet.txt","r").read()
    # 变成小写
    txt = txt.lower()
    # 处理特殊符号
    for ch in '''!"$%&()*+,-./;:<=>?@[\\]^_{|}~''\n\t ''':
          txt = txt.replace(ch," ")
    return txt
```

```python
def tongji(text_file):
    with open(text_file, 'r', encoding='utf-8') as file:
        text = file.read()

    # 使用nltk进行单词分词
    words = word_tokenize(text)

    # 统计词频
    word_freq = FreqDist(words)
    
    # 计算词频
    total_words = len(words)
    freq_dist = [(word, freq/total_words) for word, freq in word_freq.items()]

    # 获取词频最高的前20%关键词
    sorted_freq_dist = sorted(freq_dist, key=lambda x: x[1], reverse=True)
    top_20_percent = sorted_freq_dist[:int(0.2 * len(sorted_freq_dist))]

    return top_20_percent
```

```python
result = tongji('最新-正在.txt')
print(result)
```
