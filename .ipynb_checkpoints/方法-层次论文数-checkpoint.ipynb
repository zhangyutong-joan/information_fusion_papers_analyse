```python
import numpy as np
import pandas as pd
```

```python
def readxls(path):
    df=pd.read_excel(path)
    df.dropna(subset=["Keywords Plus"],inplace=True) # 删缺失的
    col=["Article Title","Keywords Plus","Abstract"]
    return df[col]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def my_plot(x_name,y_value):
    fig,ax=plt.subplots(figsize=(10, 7))
    ax.bar(x=x_name,height=y_value)
    xticks = ax.get_xticks()
    for i in range(len(y_value)):
        xy = (xticks[i], y_value[i] * 1.02)
        s = str(y_value[i])
        ax.annotate(
            text=s,  # 要添加的文本
            xy=xy,  # 将文本添加到哪个位置
            fontsize=12,  # 标签大小
            color="blue",  # 标签颜色
            ha="center",  # 水平对齐
            va="baseline"  # 垂直对齐
        )
```

```python
df_vt=readxls("./方法xls/vote-近5年-infoFusion.xlsx")
df_bs=readxls("./方法xls/Bayesian-近5年-infoFusion.xlsx")
df_st=readxls("./方法xls/stacking-近5年-infoFusion.xlsx")
df_ds=readxls("./方法xls/Dempster-Shafer-近5年-infoFusion.xlsx")
```

```python
ls={'vote':len(df_vt),'Bayesian':len(df_bs),'stacking':len(df_st),'D-S':len(df_ds)}
ls
```

```python
my_plot(list(ls.keys()),list(ls.values()))
```

```python
df_data=readxls("./层次xls/data-level fusion-近5年-infoFusion.xls")
df_fea=readxls("./层次xls/feature-level fusion-近5年-infoFusion.xls")
df_dec=readxls("./层次xls/Decision-Level Fusion-近5年-infoFusion.xls")
df_md=readxls("./层次xls/model fusion-近5年-infoFusion.xls")
```

```python
ls={'data-level':len(df_data),'feature-level':len(df_fea),'decision-level':len(df_dec),'model':len(df_md)}
ls
```

```python
my_plot(list(ls.keys()),list(ls.values()))
```
