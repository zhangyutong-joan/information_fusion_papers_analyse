```python
import numpy as np
import pandas as pd
```

```python
def readxls(path):
    df=pd.read_excel(path)
    df.dropna(subset=["Keywords Plus"],inplace=True) # 删缺失的
    col=["Article Title","Keywords Plus","Abstract","Publication Type"
         ,"Cited Reference Count","Publisher","Publication Year"]
    return df[col]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def my_plot(x_name,y_value):
    fig,ax=plt.subplots(figsize=(20, 7))
    ax.bar(x=x_name,height=y_value)
    xticks = ax.get_xticks()
    for i in range(len(y_value)):
        xy = (xticks[i], y_value[i] * 1.02)
        s = str(y_value[i])
        ax.annotate(
            text=s,  # 要添加的文本
            xy=xy,  # 将文本添加到哪个位置
            fontsize=12,  # 标签大小
            color="blue",  # 标签颜色
            ha="center",  # 水平对齐
            va="baseline"  # 垂直对齐
        )
```

```python
df=readxls("近5年-infoFusion-文章或会议文章.xls")
df.head()
```

```python
ls=df["Keywords Plus"].to_list()
ls[0]
```

```python
# 词频统计
fenci=[]
ls2={}
for value in ls:
    x=value.split('; ')
#     print(x)
    for k in x:
        fenci.append(k)
for xx in fenci:

    if xx in list(ls2.keys()):
        ls2[xx]=ls2[xx]+1
    else:
        ls2[xx]=1
        
ls2
```

```python
ls2_sort = sorted(ls2.items(),  key=lambda d: d[1], reverse=True)
ls2_sort
```

```python
# 出现超过1次的
ls2_name=[ls2_sort[i][0] for i in range(len(ls2_sort)) if ls2_sort[i][1]>1]
ls2_value=[ls2_sort[i][1] for i in range(len(ls2_sort)) if ls2_sort[i][1]>1]
len(ls2_value)
```

```python
ls3_name=ls2_name[10:20]
ls3_value=ls2_value[10:20]
ls3_name
```

```python
my_plot(ls3_name,ls3_value)
```

存一下

```python
dic={}
for i in range(len(ls2_name)):
    dic[ls2_name[i]]=ls2_value[i]
    
df_res=pd.DataFrame([dic])
df_res.head()
```

```python
df_res.to_excel("关键词统计-近5年-infoFusion-文章或会议文章.xlsx")
```
