```python
import numpy as np
import pandas as pd
```

```python
def readxls(path):
    df=pd.read_excel(path)
    col=["Article Title","Keywords Plus","Abstract"]
    return df[col]
```

```python
df_vt=readxls("./方法xls/vote-近5年-infoFusion.xlsx")
df_bs=readxls("./方法xls/Bayesian-近5年-infoFusion.xlsx")
df_st=readxls("./方法xls/stacking-近5年-infoFusion.xlsx")
df_ds=readxls("./方法xls/Dempster-Shafer-近5年-infoFusion.xlsx")
```

1. 比较论文数

```python
ls={'vote':len(df_vt),'Bayesian':len(df_bs),'stacking':len(df_st),'D-S':len(df_ds)}
ls
```

```python
fig,ax=plt.subplots(figsize=(10, 7))
ax.bar(x=list(ls.keys()),height=list(ls.values()))
xticks = ax.get_xticks()
v=list(ls.values())
for i in range(len(v)):
    xy = (xticks[i], v[i] * 1.02)
    s = str(v[i])
    ax.annotate(
        text=s,  # 要添加的文本
        xy=xy,  # 将文本添加到哪个位置
        fontsize=12,  # 标签大小
        color="blue",  # 标签颜色
        ha="center",  # 水平对齐
        va="baseline"  # 垂直对齐
    )


```

2.研究热点分析

```python
import math
df_vt.dropna(inplace=True)
df_bs.dropna(inplace=True)
df_st.dropna(inplace=True)
df_ds.dropna(inplace=True)
```

```python
# keywords取并集
ls2={'vote':df_vt["Keywords Plus"],'Bayesian':df_bs["Keywords Plus"],
     'stacking':df_st["Keywords Plus"],'D-S':df_ds["Keywords Plus"]}
fenci=[]
ls3={}
for key,value in ls2.items():
    x=value.str.split('; ')
    for k in x.values:
        fenci.append(k)
for x in fenci:
    for xx in x:
        if xx in list(ls3.keys()):
            ls3[xx]=ls3[xx]+1
        else:
            ls3[xx]=1
        
ls3
        
        
        
#     ls3=set(ls3)|set(x)
# ls3=ls3[1:]
# ls3[0]
```

```python
ls3_sort = sorted(ls3.items(),  key=lambda d: d[1], reverse=True)
ls3_sort
```

```python
len(ls3_sort)
```

```python
# 出现超过1次的
ls4_name=[ls3_sort[i][0] for i in range(len(ls3_sort)) if ls3_sort[i][1]>1]
len(ls4_name)
```

```python
ls4_value=[ls3_sort[i][1] for i in range(len(ls3_sort)) if ls3_sort[i][1]>1]
```

```python
ls5_name=ls4_name[:10]
ls5_value=ls4_value[:10]
```

```python
fig,ax=plt.subplots(figsize=(20, 7))
ax.bar(x=ls5_name,height=ls5_value)
xticks = ax.get_xticks()
for i in range(len(ls5_value)):
    xy = (xticks[i], ls5_value[i] * 1.02)
    s = str(ls5_value[i])
    ax.annotate(
        text=s,  # 要添加的文本
        xy=xy,  # 将文本添加到哪个位置
        fontsize=12,  # 标签大小
        color="blue",  # 标签颜色
        ha="center",  # 水平对齐
        va="baseline"  # 垂直对齐
    )
```
