```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt


%matplotlib inline
def tongji(text_file):
    with open(text_file, 'r') as file:
        txt = file.read().lower()
        # 处理特殊符号
        for ch in '''!"$%&()*+,-./;:<=>?@[\\]^_{|}~''\n\t ''':
            text = txt.replace(ch," ")

    
    # 使用nltk库进行词频统计
    words0 = nltk.word_tokenize(text)
    words=words0.copy()
    
    for word in words0:
        if word in ["to","the","and","in","of","for","we","with","data","that","by","on","a","is",".",",","(",")",";","are","based","from","this","each","its","an"
                   ,"two"
                   
                   
                   ]:
            words.remove(word)
    
    
    freq_dist = nltk.FreqDist(words)
    print(freq_dist)
    # 生成词云
    wordcloud = WordCloud().generate_from_frequencies(freq_dist)

    # 显示词云
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    
    freq_dist1 = [(word, freq) for word, freq in freq_dist.items()]
    # 获取词频最高的前20%关键词
    sorted_freq_dist = sorted(freq_dist1, key=lambda x: x[1], reverse=True)
    top_20_percent = sorted_freq_dist[:15]
    
#     bar(x=x, height=top_20_percent)
    print(top_20_percent)
    
    x1=[word for word, freq in freq_dist.items()]
    x1=sorted(x1, reverse=True)
    y1=[freq for word, freq in freq_dist.items()]
    y1=sorted(y1, reverse=True)
    print(x1[:5])
    
    plt.bar(x=x1[:5],height=y1[:5])
    
    
```

```python
tongji('最新-正在.txt')
```
